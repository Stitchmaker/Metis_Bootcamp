```python
import pickle
import datetime
from dateutil.parser import parse

import pandas as pd
```

```python
# read data from pickle fil
with open ('all_output.pkl', 'rb') as fp:
    all_output = pickle.load(fp)
```

## Create initial DataFrame

```python
# create a DataFrame 

df = pd.DataFrame(columns=['date','headline','snippet'])

for x in all_output:
    print(f'For month {x[0][4:]} in {x[0][:4]} there were {len(x[1])} articles')
    for i in range(len(x[1])):
        row = x[1][i]
        #print('SNIPPET: ',row['snippet'])
        #print('HEADLINE: ',row['headline'])
        #print('DATE:', row['date'], type(row['date']))
        df1  = pd.DataFrame(data=row, index=[0])
        df1.drop(columns=['date'],inplace=True)
        date = parse(row['date'])
        #df1['datetime'] = date
        df1['date'] = date.date()
        df = pd.concat([df, df1],ignore_index=True)

#print(df)  
```

```python
# write data to pickle file
with open('all_data.pkl', 'wb') as fp:
    pickle.dump(df, fp)
```

```python
import nltk
from nltk.tokenize import sent_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer

# gensim
from gensim import corpora, models, similarities, matutils
# sklearn
from sklearn import datasets
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.cluster import KMeans
# logging for gensim (set to INFO)
import logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)

#nltk.download()
```

```python
tf_vectorizer = CountVectorizer(
            max_df = 0.95, # leave out words that occur in more than 95% of docs
            min_df = 2,    # leave out if occurs less than twice
            ngram_range=(1,3),
            stop_words='english')
#tf_vectorizer.fit(df2.snippet)
#counts = tf_vectorizer.transform(df2.snippet).transpose()

tf = tf_vectorizer.fit_transform(df2.snippet)
```

```python
tf.shape
```

```python
corpus = matutils.Sparse2Corpus(tf)
```

```python
len(corpus)
```

```python
tf_vectorizer.vocabulary_
```

```python
id2word = dict((v, k) for k, v in tf_vectorizer.vocabulary_.items())
len(id2word)
```

```python
id2word
```

```python
# Create lda model (equivalent to "fit" in sklearn)
lda = models.LdaModel(corpus=corpus, num_topics=5, minimum_probability=0.03, id2word=id2word, passes=10)
```

```python
lda.print_topics()
```

```python
# Transform the docs from the word space to the topic space (like "transform" in sklearn)
lda_corpus = lda[corpus]
lda_corpus
```

```python
# Store the documents' topic vectors in a list so we can take a peak
lda_docs = [doc for doc in lda_corpus]
```

```python
# Check out the document vectors in the topic space for the first 5 documents
lda_docs[:15]
```

```python
# we can quantify the 'fit' of our model, to compare with other corpora, etc.
lda.log_perplexity(corpus)
```

```python
tf_vectorizer = CountVectorizer(
            max_df = 0.95, # leave out words that occur in more than 95% of docs
            min_df = 2,    # leave out if occurs less than twice
            ngram_range=(1,3),
            stop_words='english')
#tf_vectorizer.fit(df2.snippet)
#counts = tf_vectorizer.transform(df2.snippet).transpose()

dtm_tf = tf_vectorizer.fit_transform(df2.snippet)
print(dtm_tf.shape)
```

```python
tfidf_vectorizer = TfidfVectorizer(**tf_vectorizer.get_params())
dtm_tfidf = tfidf_vectorizer.fit_transform(df2.snippet)
print(dtm_tfidf.shape)
```

```python
from sklearn.decomposition import LatentDirichletAllocation
# for TF DTM
lda_tf = LatentDirichletAllocation(n_topics=20, random_state=0)
lda_tf.fit(dtm_tf)
# for TFIDF DTM
lda_tfidf = LatentDirichletAllocation(n_topics=20, random_state=0)
lda_tfidf.fit(dtm_tfidf)
```
