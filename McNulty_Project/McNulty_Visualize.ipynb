```python
# How may flights are on-time (0) and how many delayed (1){}
FlightData['Delay'].value_counts()
```

```python
DelayCount = {}
OnTimeCount = {}
for index,row in FlightData.iterrows():
    date = row['FL_DATE']
    if row['Delay']:
        if date in DelayCount:
            DelayCount[date] += 1
        else:
            DelayCount[date] = 1
    else:
        if date in OnTimeCount:
            OnTimeCount[date] += 1
        else:
            OnTimeCount[date] = 1
```

```python
plt.bar(DelayCount.keys(), DelayCount.values(), color='r');
```
